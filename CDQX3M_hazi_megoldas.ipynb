```sql
SELECT MIN(EGYSEGAR) AS min_egysegar, 
       MAX(EGYSEGAR) AS max_egysegar, 
       AVG(EGYSEGAR) AS atlag_egysegar, 
       STDEV(EGYSEGAR) AS szoras_egysegar
FROM Rendeles_tetel
WHERE EGYSEGAR IS NOT NULL
```

```sql
SELECT MIN(MENNYISEG) AS min_mennyiseg, 
       MAX(MENNYISEG) AS max_mennyiseg, 
       AVG(MENNYISEG) AS atlag_mennyiseg, 
       STDEV(MENNYISEG) AS szoras_mennyiseg
FROM Rendeles_tetel
WHERE MENNYISEG IS NOT NULL
```

```sql
SELECT TOP 10
    FIZ_MOD AS 'fizetsi md', 
    COUNT(*) 'sszesen'
FROM Rendeles
WHERE FIZ_MOD IS NOT null
GROUP BY FIZ_MOD
ORDER BY COUNT(*) desc
```

```sql
SELECT DATEDIFF(day, REND_DATUM, SZALL_DATUM) as "Szllts ideje (nap)", COUNT(*) as "Darab"
FROM Rendeles
GROUP BY DATEDIFF(day, REND_DATUM, SZALL_DATUM)
ORDER BY COUNT(*) DESC
```

```sql
SELECT * 
FROM Termek
WHERE TERMEKKOD IN (
    SELECT TOP (10) TERMEKKOD
    FROM Termek
    ORDER BY NEWID())
```

```sql
SELECT TOP (10) [LOGIN], SZALL_CIM, REND_DATUM
FROM Rendeles
GROUP BY [LOGIN], SZALL_CIM, REND_DATUM
ORDER BY NEWID()
```
